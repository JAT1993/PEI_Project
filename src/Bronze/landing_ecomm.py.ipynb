```python
from pyspark.sql import DataFrame
def clean_column_spaces(df: DataFrame) -> DataFrame:

    """
    Replaces spaces in DataFrame column names with underscores.
    """
    # Create a mapping dictionary: {"old name": "new_name"}
    rename_mapping = {
        col: col.replace(" ", "") 
        for col in df.columns 
        if " " in col
    }
    
    # Apply all renames efficiently in a single operation
    return df.withColumnsRenamed(rename_mapping)
```

Customer Table creation

```python
def excel_read(excel_path):
    df_excel=spark.read.format("excel")\
        .option("headerRows", 1) \
        .option("inferSchema","true")\
        .load(excel_path)
    return clean_column_spaces(df_excel)   
```

```python
def excel_write(table_name,excel_path):
    df_excel_clean=excel_read(excel_path)
    return df_excel_clean.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable(table_name)
```

Product table creation

```python
def csv_read(csv_path):
    df_product=spark.read.format("csv")\
        .option("inferSchema","true")\
        .option("header","true")\
        .load(csv_path)
    return clean_column_spaces(df_product)
```

```python
def csv_write(table_name,csv_path):
    df_csv_clean=csv_read(csv_path)
    return df_csv_clean.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable(table_name)
```

```python
def json_read(json_path):
    df_orders=spark.read.format("json")\
        .option("multiLine","true")\
        .load(json_path)
    return clean_column_spaces(df_orders)
```
